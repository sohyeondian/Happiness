```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import seaborn as sns
```

```python
hp = pd.read_csv('data/2020.csv')
hp.info()
```

```python
hp.head()
```

```python
data = hp.drop(['upperwhisker', 'lowerwhisker', 'Standard error of ladder score', 
                'Ladder score in Dystopia', 'Explained by: Log GDP per capita', 
                'Explained by: Social support', 'Explained by: Healthy life expectancy', 
                'Explained by: Freedom to make life choices', 'Explained by: Generosity', 
                'Explained by: Perceptions of corruption'], axis=1)
data.head()
```

```python
plt.rcParams['figure.figsize'] = (12,8)
sns.heatmap(data.corr(), cmap = 'copper', annot = True)

plt.show()
```

```python
data[['Country name','Ladder score']].head(10)
```

```python
# 대륙의 나라 수 확인
regions = data['Regional indicator'].value_counts()
```

```python
sns.set(style="darkgrid")
sns.barplot(regions.values, regions.index, alpha=0.9)
plt.xlabel('Frequency', fontsize=12)
plt.ylabel('Regions', fontsize=12)
plt.show()
```

```python
# 대륙별 점수 평균 - 순서대로
reg_data = data.groupby('Regional indicator')['Ladder score'].mean()
reg_data.sort_values(ascending=False, inplace=True)
reg_data
```

```python
sns.set(style="darkgrid")
sns.barplot(reg_data.values.tolist(), reg_data.index.tolist(), alpha=0.9)
plt.title('MEAN Score by Continent', fontsize=15)
plt.xlabel('Score', fontsize=12)
plt.ylabel('Regions', fontsize=12)
plt.show()
```
