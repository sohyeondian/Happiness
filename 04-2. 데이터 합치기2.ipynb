```python
def add_reg(df, col, reg_data):
    match_continent = {}
    for name in df[col]:
        tmp = reg_data[reg_data['Country']==name]['Region'].tolist()
        if len(tmp):
            match_continent[name] = tmp[0]
    
    match_df = pd.DataFrame.from_dict(match_continent, 
                                      orient='index', 
                                      columns=['Region'])
    match_df = match_df.reset_index().rename(columns={'index':col})
    merged_df = pd.merge(df, match_df, on=[col], how='left')
    return merged_df
```
